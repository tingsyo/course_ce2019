上週我們提供了整理好的 EPA 測站資料，這週的主題是嘗試用 PM2.5 的測站觀測資料，來定義所謂的「高污染事件」。

首先，我們拿 2015 年的資來做測試：

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# For Chinese font
from matplotlib.font_manager import FontProperties
font = FontProperties(fname="../data/NotoSansCJKtc-Regular.otf", size=10)

data = pd.read_csv('../data/pm25_2015.csv')
```

首先，我們可以看一下所有觀測的 histogram，了解一下 PM2.5 觀測濃度的概況：

```python
pm25 = data.iloc[:,2:]
pm25.index = data['date']+':'+data['hour']
pm25_summary = pm25.describe()
pm25list = pm25.values.flatten()[~np.isnan(pm25.values.flatten())]
pm25_hist = np.histogram(pm25list, bins=np.arange(0,350,50))
# Show plot
fig, axs = plt.subplots(1, 2, figsize=(9, 3))
axs[0].hist(pm25list, bins=np.arange(0,350,25))
axs[0].set_title('Histogram of all EPA PM2.5 observations in 2015')
axs[1].hist(pm25list, bins=np.arange(0,350,25), density=True, cumulative=True)
axs[1].set_title('Cumulative Probability')
plt.tight_layout()
plt.show()
#
print('Cumulative probability: ')
print(np.cumsum(pm25_hist[0]/sum(pm25_hist[0])))
print(pm25_hist[1])
```

我們可以看到，PM2.5 濃度高於 50, 100, 150 ppm 的機率大致是 7.7%, 0.18%, 和 0.005%。對照環保署[空氣品質指標](https://taqm.epa.gov.tw/taqm/tw/b0201.aspx)的定義，如果純粹考慮 PM2.5，「對敏感族群不健康」、「對所有族群不健康」、「非常不健康」、「危害」的標準，分別對應到的是 PM2.5 24小時平均 35.5, 54.5, 150.4, 250.4 ppm。

因此，我們可以先對資料集做24小時的移動平均（moving average），觀察各種狀態的發生情況：

```python
pm25_ma24 = pm25.rolling(24).mean()
print('data dimension: '+str(pm25_ma24.shape))
# Plot
plt.figure(figsize=(16,18))
plt.contourf(pm25_ma24, levels=[0, 35.5, 54.5, 150.4, 250.4], cmap='Greys')
plt.xticks(np.arange(pm25_ma24.shape[1]), pm25_ma24.keys(), rotation=90, fontproperties=font)
plt.yticks(np.arange(0,(pm25_ma24.shape[0]),720), pm25_ma24.index[np.arange(0,(pm25_ma24.shape[0]),720)])
plt.colorbar()
plt.tight_layout()
plt.show()
```

由圖中可以看到，比較高濃度的事件通常發生在10月～4月之間，也比較集中在中南部的測站。

## 設計「嚴重污染」的指標

接下來，我們設定一個 threshold，把所有測站的值疊加起來，計算一定時間的總量，作為一個指標，可以得到一個台灣地區單一直的時間序列。

```python
# Function to calculae a single time series of air-polution-index
def apindex(pm25, threshold=35.5, period=120):
    tmp = (pm25>=threshold)*1
    tmp = tmp.sum(axis=1)
    tmp = tmp.rolling(period).sum()
    return(tmp/(period*pm25.shape[1]))

# Set xticks
xs = np.arange(0, pm25_ma24.shape[0], 720)
# Test the period
ap1 = apindex(pm25_ma24, period=24)
ap3 = apindex(pm25_ma24, period=72)
ap5 = apindex(pm25_ma24, period=120)
ap7 = apindex(pm25_ma24, period=168)
ap10 = apindex(pm25_ma24, period=240)
#
plt.figure(figsize=(12,8))
plt.plot(list(ap1), 'k-', label='1day')
plt.plot(list(ap3), 'b--', label='3day')
plt.plot(list(ap5), 'r-.', label='5day')
plt.plot(list(ap7), 'g--', label='7day')
plt.plot(list(ap10), 'y-.', label='10day')
plt.xticks(xs, pm25_ma24.index[xs], rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
aplabel = ['Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']
# Test the threshold
ap1 = apindex(pm25_ma24, threshold=35.5)
ap2 = apindex(pm25_ma24, threshold=54.5)
ap3 = apindex(pm25_ma24, threshold=150.4)
ap4 = apindex(pm25_ma24, threshold=250.4)
#
plt.figure(figsize=(12,8))
plt.plot(list(ap1), 'k-', label=aplabel[0])
plt.plot(list(ap2), 'b--', label=aplabel[1])
plt.plot(list(ap3), 'r-.', label=aplabel[2])
plt.plot(list(ap4), 'g--', label=aplabel[3])
plt.xticks(xs, pm25_ma24.index[xs], rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

從以上的測試結果來看，累計的週期長度對整體趨勢的想不大，只有平滑的效果；而 threshold 選在 AQI >100 似乎是比較好的選擇，再往上事件發生的機率太低；此外，每個峰值涵蓋的範圍大約都在10天以內（每個時間標記間隔30天）。我們接下來將 24 小時移動平均的時間序列，對所有測站加總，然後進一步拆成 10 天的 time window，然後進行 cluster analysis。

```python
# Function to compute summed air-pollution-index and cut the whole series with specified time window
def api_timeseries(pm25, threshold=54.5, window=240):
    tmp = (pm25>=threshold)*1
    tmp = tmp.sum(axis=1)
    ts = {}
    for i in np.arange(23+window, len(tmp), 24):
        #print('Index: '+str(i-window)+' ~ '+str(i))
        ts[tmp.index[i]] = list(tmp[i-window:i])
    df = pd.DataFrame(ts)
    return(df)

apts = api_timeseries(pm25_ma24)
apts.shape
```
